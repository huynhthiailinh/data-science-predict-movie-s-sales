```python
import requests
from requests_html import HTMLSession
from bs4 import BeautifulSoup
import pandas as pd
```

# crawl chi tiết xe

```python
# Khỏi tạo list carDetail để lưu dữ liệu crawl được
carDetail=list()
```

```python
def get_detail(id,price,name,url):
    # gửi HTTP request đến url
    Oto_link='https://bonbanh.com/'+ url
    session = HTMLSession()
    req=session.get(Oto_link)
    element = req.html.html
        
    soup = BeautifulSoup(element,"html.parser")
    # Lấy ra cột thứ nhất chứa dữ liệu
    data_column= soup.findAll("div",{"class":"col"})[0]
    
    # Lấy ra các dữ liệu từ các thẻ 
    origin=data_column.findAll("span",{"class":"inp"})[0].text
    condition=data_column.findAll("span",{"class":"inp"})[1].text
    carType=data_column.findAll("span",{"class":"inp"})[2].text
    km=data_column.findAll("span",{"class":"inp"})[3].text
    carColor=data_column.findAll("span",{"class":"inp"})[4].text
    interiorColor = data_column.findAll("span",class_='inp')[5].text
    carDoor = data_column.findAll("span",class_='inp')[6].text
    carSeat=data_column.findAll("span",{"class":"inp"})[7].text

    # Lấy ra cột thứ hai chứa dữ liệu
    data_column= soup.findAll("div",{"class":"col"})[1]

    # Lấy ra các dữ liệu từ các thẻ 
    engine=''.join(data_column.findAll("span",{"class":"inp"})[0].text.split())
    gear=data_column.findAll("span",{"class":"inp"})[2].text
    wheelDrive=data_column.findAll("span",{"class":"inp"})[3].text

    # Thêm các giá trị crawl được vào carDetail với id tương ứng
    carDetail.append({'id':id,'car':name,'price':price,' origin': origin,'condition':condition,'carType':carType,'km':km,'carColor':carColor,
                      'interiorColor':interiorColor,'carDoor':carDoor,'carSeat':carSeat,'engine':engine,
                      'gear':gear,'wheelDrive':wheelDrive})
```

```python
session = HTMLSession()
# Số trang crawl
n=5
#  vòng lặp để truy cập từng trang bài đăng
for k in range(1,n):
    # gửi HTTP request đến url
    access_homepage = "https://bonbanh.com/oto/page,"+str(k)
    req=session.get(access_homepage)
    element = req.html.html
    print(access_homepage)
    
    # Sử dụng BeautifulSoup để lấy ra list bài đăng
    soup = BeautifulSoup(element,'html.parser')
    product_list= soup.findAll("li",{"class":"car-item"})
    # vòng for lặp từng bài đăng có trong list của 1 trang
    for product in product_list:
        # Lấy ra các giữ liệu cần thiết
        id=product.find("span",{"class":"car_code"}).text.strip()
        url=product.find("a").get('href')
        Name=product.find("div",{"class":"cb2_02"}).text.strip()
        Price=product.find("b",{"itemprop":"price"}).get("content")
        print(id)
        print(url)
        print(Name)
        print(Price)
        # gọi hàm get_detail để truy cập và lấy ra thông tin chi tiết của bài đăng
        get_detail(id,Price,Name,url)
       
```

```python
# Ghi dữ liệu vừa crawl được vào file raw_data.csv
data = pd.DataFrame.from_dict(carDetail)
print(data)
data.to_csv('raw_data.csv', index=False,encoding="utf-8-sig",
                  columns=['id','car','price',' origin','condition','carType','km','carColor',
                      'interiorColor','carDoor','carSeat','engine',
                      'gear','wheelDrive'])

```
